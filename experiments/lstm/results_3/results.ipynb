```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
def read_log(fname):
    file = open(fname, 'r')
    x = []
    accuracy = []
    next(file)
    for line in file.readlines():
        it, current, target, prediction, correct = line.replace('\n','').split(',')
        x.append(float(it))
        accuracy.append(float(correct))
    return x, accuracy
```

```python
def moving_average(a, n):
    moving_average = []
    for i in range(len(a)):
        start = max(0, i - n)
        values = a[start:i+1]
        moving_average.append(np.sum(values) / float(len(values)))
    return np.array(moving_average)
```

```python
logs = [i for i in os.listdir('.') if '.log' in i]
```

```python
averages = []
x = None
for log in logs:
    x, acc = read_log(log)
    averages.append(moving_average(acc, min(len(acc), 200)))
plt.plot(x, np.mean(averages, axis=0), label='online-LSTM')
plt.fill_between(x, (np.mean(averages, axis=0)+np.std(averages, axis=0)), (np.mean(averages, axis=0)-np.std(averages, axis=0)), color=(0.1, 0.1, 0.9, 0.1))
plt.plot([10000, 10000], [0, 1], linestyle='--', color='gray')
plt.ylim(0, 1.0)
plt.xlim(0, 20000)
plt.title('LSTM-online')
plt.ylabel('Accuracy')
plt.xlabel('Iteration')
plt.savefig('output.png', dpi=100)
plt.show()
```
