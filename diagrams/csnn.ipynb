```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.gca()
edgecolor = (0, 0, 0, 0.9)
neuroncolor = (0, 0, 0)
ax.text(0.4, 1, 'CSNN Network Model', size=20, ha='center', va='center')

for i in range(0, 10):
    ax.add_artist(plt.Line2D([-0.01, 0.1], [0.5, (0.1 + 0.09*i)], c=edgecolor, linewidth=0.7))
    ax.add_artist(plt.Circle((0.1, 0.1 + 0.09*i), 0.03, color='white', ec=neuroncolor, linewidth=1, zorder=4))
    
ax.text(0.1, 0, '10 Input with\n25 Neurons each', size=15, ha='center', va='center')

colors = [(0, 0, 0, 0.4)]*10
for i in range(0, 10):
    ax.add_artist(plt.Rectangle((0.4, 0.07 + 0.09*i), 0.05, 0.06, fill=False, color=colors[i]))
    for j in range(0, 10):
        if i != j:
            continue
        ax.add_artist(plt.Line2D([0.1, 0.4], [(0.1+0.09*j), (0.1+0.09*(i))], c=edgecolor, linewidth=0.7))
    
    for j in range(i, 10):
        ar = patches.FancyArrowPatch((0.4+0.05,0.07 + 0.09*i + 0.03), (0.4+0.05,0.07 + 0.09*j + 0.03), connectionstyle='arc3,rad=.4', color=colors[i])
        ax.add_patch(ar)

        
ax.text(0.48, 0, '10 Columns with\n100 Neurons each', size=15, ha='center', va='center')

for i in range(0, 10):
    if i > 0:
        ax.text(-0.08, 0.65+i*0.03, 'Character$_{t-'+str(i)+'}$', size=10, ha='center', va='center')
    else:
        ax.text(-0.08, 0.65+i*0.03, 'Character$_t$', size=10, ha='center', va='center')


ax.add_artist(plt.Arrow(-0.08, 0.625, 0, -0.07, fill=False, width=0.04))
ax.add_artist(plt.Rectangle((-0.16, 0.465), 0.15, 0.08, fill=False))
ax.text(-0.08, 0.5, 'Encoder', size=12, ha='center', va='center')

ax.add_artist(plt.Arrow(0.67, 0.5, 0.07, 0, fill=False, width=0.04))
ax.add_artist(plt.Rectangle((0.8, 0.46), 0.16, 0.08, fill=False))
ax.text(0.885, 0.5, 'KNN', size=12, ha='center', va='center')
ax.add_artist(plt.Arrow(0.885, 0.445, 0, -0.07, fill=False, width=0.04))
ax.text(0.885, 0.35, 'Character$_{t+1}$', size=10, ha='center', va='center')

ax.text(0.885, 0, 'Spiking Activity\nClassifier', size=15, ha='center', va='center')

ax.axis('off')
ax.set_aspect('equal', adjustable='box')
plt.xlim(-0.2, 1.1)
plt.ylim(0, 1)
plt.savefig('csnn.pdf', dpi=200)
plt.savefig('csnn.png', dpi=200)
plt.savefig('csnn.svg', type='svg')
plt.show()
```
