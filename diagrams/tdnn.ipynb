```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.gca()
edgecolor = (0, 0, 0, 0.9)
neuroncolor = (0, 0, 0)
neuroncolor2 = (1, 0.5, 0)
ax.text(0.4, 1, 'TD(S)NN Network Model', size=20, ha='center', va='center')

for i in range(0, 10):
    ax.add_artist(plt.Line2D([-0.01, 0.1], [0.5, (0.1 + 0.09*i)], c=edgecolor, linewidth=0.7))
    ax.add_artist(plt.Circle((0.1, 0.1 + 0.09*i), 0.03, color='white', ec=neuroncolor, linewidth=1, zorder=4))
    
ax.text(0.1, 0, '250 Input', size=15, ha='center', va='center')
for i in range(0, 8):
    ax.add_artist(plt.Circle((0.4, 0.19 + 0.09*i), 0.03, color='white', ec=neuroncolor, linewidth=1, zorder=4))
    for j in range(0, 10):
        ax.add_artist(plt.Line2D([0.1, 0.4], [(0.1+0.09*j), (0.1+0.09*(i+1))], c=edgecolor, linewidth=0.7))
ax.text(0.4, 0, '200 Hidden', size=15, ha='center', va='center')

for j in range(0, 8):
    ax.add_artist(plt.Line2D([0.4, 0.7], [(0.1+0.09*(j+1)), 0.55], c=edgecolor, linewidth=0.7))
ax.add_artist(plt.Circle((0.7, 0.55), 0.03, color='white', ec='black', linewidth=1, zorder=4))


for j in range(0, 8):
    ax.add_artist(plt.Line2D([0.4, 0.7], [(0.1+0.09*(j+1)), 0.45], c=edgecolor, linewidth=0.7, linestyle='--'))
ax.add_artist(plt.Circle((0.7, 0.45), 0.03, color=neuroncolor, ec='black', linewidth=1, zorder=4))


ax.add_artist(plt.Line2D([0.7, 0.805], [0.45, 0.5], c=edgecolor, linewidth=0.7))
ax.add_artist(plt.Line2D([0.7, 0.805], [0.55, 0.5], c=edgecolor, linewidth=0.7))
ax.text(0.75, 0, '25 Output', size=15, ha='center', va='center')


for i in range(0, 10):
    if i > 0:
        ax.text(-0.08, 0.65+i*0.03, 'Character$_{t-'+str(i)+'}$', size=10, ha='center', va='center')
    else:
        ax.text(-0.08, 0.65+i*0.03, 'Character$_t$', size=10, ha='center', va='center')


ax.add_artist(plt.Arrow(-0.08, 0.625, 0, -0.07, fill=False, width=0.04))
ax.add_artist(plt.Rectangle((-0.16, 0.465), 0.15, 0.08, fill=False))

ax.text(-0.08, 0.5, 'Encoder', size=12, ha='center', va='center')
ax.add_artist(plt.Rectangle((0.805, 0.465), 0.15, 0.08, fill=False))

ax.text(0.88, 0.5, 'Decoder', size=12, ha='center', va='center')
ax.add_artist(plt.Arrow(0.88, 0.445, 0, -0.07, fill=False, width=0.04))
ax.text(0.88, 0.35, 'Character$_{t+1}$', size=10, ha='center', va='center')

ax.axis('off')
ax.set_aspect('equal', adjustable='box')
plt.xlim(-0.2, 1)
plt.ylim(0, 1)
plt.savefig('tdnn.pdf', dpi=200)
plt.savefig('tdnn.png', dpi=200, bbox_inches='tight')
plt.savefig('tdnn.svg', type='svg')
plt.show()
```
