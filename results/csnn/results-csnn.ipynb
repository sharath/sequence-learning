```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_log(fname):
    file = open(fname, 'r')
    next(file)
    td_x, csnn = [], []
    for line in file.readlines():
        line = line.strip('\n').split(',')
        try:
            it,target,csnn_prediction,accuracy = line
        except:
            break
        td_x.append(int(it))
        csnn.append(float(accuracy))
    return td_x, csnn
```

```python
plt.figure(figsize=(15, 8))

td_x, csnn = read_log('slurm-4855742.out')
plt.plot(td_x, csnn, label='CSNN')

td_x, csnn = read_log('slurm-4855743.out')
plt.plot(td_x, csnn, label='CSNN')

td_x, csnn = read_log('slurm-4855744.out')
plt.plot(td_x, csnn, label='CSNN')

td_x, csnn = read_log('slurm-4855745.out')
plt.plot(td_x, csnn, label='CSNN')

td_x, csnn = read_log('slurm-4855746.out')
plt.plot(td_x, csnn, label='CSNN')

plt.legend()
plt.xticks(np.linspace(0, 20000, 11))
plt.xlim(0, 20000)
plt.ylim(0, 1)
plt.grid()
plt.savefig('csnn.pdf', dpi=150)
plt.show()
```
