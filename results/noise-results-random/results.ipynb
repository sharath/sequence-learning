```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_log(fname):
    file = open(fname, 'r')
    next(file)
    td_x = []
    tdnn = []
    tdsnn = []
    noise = None
    for line in file.readlines():
        line = line.strip('\n').split(',')
        it,target,tdnn_prediction,tdsnn_prediction,training_loss,noise_level = line
        td_x.append(int(it))
        tdnn.append(int(target == tdnn_prediction))
        tdsnn.append(int(target == tdsnn_prediction))
        noise = float(noise_level)
    return td_x, tdnn, tdsnn, noise
```

```python
def moving_average(a, n=100):
    moving_average = []
    for i in range(len(a)):
        start = max(0, i - n)
        values = a[start:i+1]
        moving_average.append(np.sum(values) / float(len(values)))
    return np.array(moving_average)
```

```python
def mean_accuracy(fname, s=7000, e=10000):
    file = open(fname,'r')
    next(file)
    td_x = []
    tdnn = []
    tdsnn = []
    noise_level = None
    for line in file.readlines():
        it,target,tdnn_prediction,tdsnn_prediction,training_loss,noise_level = line.strip('\n').split(',')
        td_x.append(int(it))
        tdnn.append(int(target == tdnn_prediction))
        tdsnn.append(int(target == tdsnn_prediction))
        noise_level = float(noise_level)
    tdnn = moving_average(tdnn)
    tdsnn = moving_average(tdsnn)
    tdnn_ = []
    tdsnn_ = []
    for idx, it in enumerate(td_x):
        if it < s or it > e:
            continue
        tdnn_.append(tdnn[idx])
        tdsnn_.append(tdsnn[idx])
    return np.mean(tdnn_), np.mean(tdsnn_), noise_level
```

```python
start = 4845303
n_x = []
tdnn_avg = []
tdsnn_avg = []
while start <= 4845322:
    fname = f'slurm-{start}.out'
    tdnn, tdsnn, noise = mean_accuracy(fname)
    n_x.append(noise)
    tdnn_avg.append(tdnn)
    tdsnn_avg.append(tdsnn)
    start += 1

plt.figure(figsize=(10, 8))
plt.plot(n_x, tdnn_avg, label='ANN', color='teal')
plt.plot(n_x, tdsnn_avg, label='SNN', color='green')
plt.title('Noise vs. Accuracy')
plt.xlabel('Noise ($\\alpha$)')
plt.ylabel('Mean Accuracy (7000 - 10000)')
plt.xlim(0, 0.2)
plt.ylim(0, 1)
plt.grid()
plt.legend()

plt.savefig('noise-vs-accuracy.pdf', dpi=150)
plt.show()
```
